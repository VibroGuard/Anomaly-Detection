```python
import matplotlib.pyplot as plt
```

```python
with open("x_data.txt", "rt") as x_data_file:
  line = x_data_file.readline()
  x_data = list(map(float, line.split()))
  # print(x_data)

  plt.subplot(2, 1, 1)
  plt.plot(x_data)  # full dataset

  plt.subplot(2, 1, 2)
  plt.plot(x_data[7500:8000]) # zoomed-in part of the dataset
```
